```python
import cv2
import csv
import cvzone
import time
from cvzone.HandTrackingModule import HandDetector
```

```python
cap = cv2.VideoCapture(1)
cap.set(3, 1280)
cap.set(4, 720)
detector = HandDetector(detectionCon=0.8)

class MCQ():
    def __init__(self,data):
        self.question=data[0]
        self.option1=data[1]
        self.option2=data[2]
        self.option3=data[3]
        self.option4=data[4]
        self.answer=int(data[5])

        self.userAns=None

    def update(self,cursor,bboxs):

        for x,bbox in enumerate(bboxs):
            x1,y1,x2,y2=bbox
            if x1<cursor[0]<x2 and y1<cursor[1]<y2:
                self.userAns=x+1
                cv2.rectangle(Frame,(x1,y1),(x2,y2),(111,125,115),cv2.FILLED)

# Importing csv file
pathCSV=r"C:\Users\sruja\Downloads\Virtual_Quiz_Questions.csv"
with open(pathCSV,newline='\n') as f:
    reader = csv.reader(f)
    dataAll = list(reader)[1:]

# Create object for each mcq
mcqList=[]
for q in dataAll:
    mcqList.append(MCQ(q))

qNo=0
qTot=len(dataAll)

while True:
    success, Frame = cap.read()
    Frame=cv2.flip(Frame, 1)
    hands, Frame = detector.findHands(Frame, flipType=False)

    if qNo<qTot:
        mcq = mcqList[qNo]
    
        Frame,bbox=cvzone.putTextRect(Frame,mcq.question,[350,100],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=50,border=5,colorB=(0,0,0))
        Frame,bbox1=cvzone.putTextRect(Frame,mcq.option1,[400,250],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=50,border=5,colorB=(0,0,0))
        Frame,bbox2=cvzone.putTextRect(Frame,mcq.option2,[800,250],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=50,border=5,colorB=(0,0,0))
        Frame,bbox3=cvzone.putTextRect(Frame,mcq.option3,[400,400],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=50,border=5,colorB=(0,0,0))
        Frame,bbox4=cvzone.putTextRect(Frame,mcq.option4,[800,400],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=50,border=5,colorB=(0,0,0))
    
        if hands:
            lmList=hands[0]['lmList']
            cursor=lmList[8]
            length, info, _=detector.findDistance(lmList[8][:2], lmList[12][:2])
            if length<30:
                mcq.update(cursor,[bbox1,bbox2,bbox3,bbox4])
                print(mcq.userAns)
                if mcq.userAns is not None:
                    time.sleep(0.3)
                    qNo+=1
                    

    else:
        score=0
        for mcq in mcqList:
            if mcq.answer == mcq.userAns:
                score+=1

        score=round((score/qTot)*100,2)
        Frame,_=cvzone.putTextRect(Frame,"Quiz Completed",[250,300],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=50,border=5,colorB=(0,0,0))
        Frame,_=cvzone.putTextRect(Frame,f'Your Score: {score}%',[700,300],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=50,border=5,colorB=(0,0,0))



    # Draw Progress Bar
    barValue=150+(950//qTot)*qNo
    cv2.rectangle(Frame,(150,600),(barValue,650),(232,184,12),cv2.FILLED)
    cv2.rectangle(Frame,(150,600),(1100,650),(232,184,12),5)
    Frame,_=cvzone.putTextRect(Frame,f'{round((qNo/qTot)*100)}%',[1130,635],2,2,colorT=(0,0,0),colorR=(232,184,12),offset=16,border=5,colorB=(232,184,12))
        

    cv2.imshow('Image', Frame)
    
    if cv2.waitKey(1)&255 ==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
